```python
# import time
import pandas as pd
import plotly.express as px
import streamlit as st
from numpy.random import default_rng as rng
# import numpy as np


print("Loading data...")
car_data = pd.read_csv('C:/Users/MonoTono2/Documents/Curso_TripleTen/Proyectos/Sprint_7/GitHub_Ejercicios/vehicles_us.csv')
# hist_button = st.button('Construir histograma')
st.write('Tabla completa del conjunto de datos de vehículos en EE.UU.:')
st.dataframe(car_data, width=1200, height=700)
st.write('Histograma del odómetro de los vehículos en EE.UU.:')
print("Data loaded successfully.")
# st.bar_chart(car_data['odometer'].value_counts().sort_index())
# hist_button = st.button('Construir histograma')
fig = px.histogram(car_data, x='odometer', nbins=100, title='Histograma del odómetro de los vehículos en EE.UU.')
st.plotly_chart(fig, use_container_width=True)

#st.bar_chart(car_data, x='model_year', y='model', x_label='Año del modelo', y_label='Modelo', color='#ffaa00', sort=True)
#st.bar_chart(car_data, x='odometer', y='type', x_label='Odometro', y_label='Tipo de vehículo', color='#ffaa00')
st.bar_chart(car_data['odometer'].value_counts(dropna=True).sort_values(ascending=True), x_label='Odómetro', y_label='Count', color="#066b1c")

st.area_chart(car_data['odometer'].value_counts().sort_index(), width="stretch", height="stretch", use_container_width=True)

# st.write('Gráfico de barras horizontales del odómetro de los vehículos en EE.UU.:')
# # Mostrar el gráfico de barras horizontales
# st.bar_chart(car_data['odometer'].value_counts(bins=100).sort_index(), horizontal=True)
```
